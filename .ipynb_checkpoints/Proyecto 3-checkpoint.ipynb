```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
```

```python
data_file = 'MaguaP.csv'
data = pd.read_csv(data_file, header = None)
```

```python
x = data.iloc[:, 0].values
y = data.iloc[:, 1].values

plt.figure()
plt.plot(x, y)
plt.show()
```

```python
import pyomo_utilities
```

```python
beta1 = pyomo_utilities.curve_polyfit(x, y, 1)
beta2 = pyomo_utilities.curve_polyfit(x, y, 2)
beta3 = pyomo_utilities.curve_polyfit(x, y, 3)
beta4 = pyomo_utilities.curve_polyfit(x, y, 4)
```

```python
yhat1 = beta1.dot(np.array([x**i for i in range(2)]))

plt.figure(figsize = (8,6))
plt.plot(x, y, '*b', label = 'datos reales')
plt.plot(x, yhat1, '-r', label = 'ajuste 1')

plt.legend(loc = 'best')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python

yhat2 = beta2.dot(np.array([x**i for i in range(3)]))



plt.figure(figsize = (8,6))
plt.plot(x, y, '*b', label = 'datos reales')

plt.plot(x, yhat2, 'oy', label = 'ajuste 2')

plt.legend(loc = 'best')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python

yhat3 = beta3.dot(np.array([x**i for i in range(4)]))



plt.figure(figsize = (8,6))
plt.plot(x, y, '*b', label = 'datos reales')

plt.plot(x, yhat3, '-k', label = 'ajuste 3')

plt.legend(loc = 'best')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python

yhat4 = beta4.dot(np.array([x**i for i in range(5)]))


plt.figure(figsize = (8,6))
plt.plot(x, y, '*b', label = 'datos reales')

plt.plot(x, yhat4, '--g', label = 'ajuste 4')
plt.legend(loc = 'best')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
yhat1 = beta1.dot(np.array([x**i for i in range(2)]))
yhat2 = beta2.dot(np.array([x**i for i in range(3)]))
yhat3 = beta3.dot(np.array([x**i for i in range(4)]))
yhat4 = beta4.dot(np.array([x**i for i in range(5)]))


plt.figure(figsize = (8,6))
plt.plot(x, y, '*b', label = 'datos reales')
plt.plot(x, yhat1, '-r', label = 'ajuste 1')
plt.plot(x, yhat2, 'oy', label = 'ajuste 2')
plt.plot(x, yhat3, '-k', label = 'ajuste 3')
plt.plot(x, yhat4, '--g', label = 'ajuste 4')
plt.legend(loc = 'best')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
ems = []
ems.append(sum((y-yhat1)**2))
ems.append(sum((y-yhat2)**2))
ems.append(sum((y-yhat3)**2))
ems.append(sum((y-yhat4)**2))

plt.figure(figsize = (8,6))
plt.plot(np.arange(4)+1, ems, '*b')
plt.xlabel('$n$')
plt.ylabel('$e_{ms}(n)$')
plt.show()
```
